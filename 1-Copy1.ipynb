# concatinate all labels

```python
labels = dict_1[b'labels'] + dict_2[b'labels'] +dict_3[b'labels']+dict_4[b'labels']+dict_5[b'labels']
```

```python
array = np.concatenate([dict_1[b'data'],dict_2[b'data'],dict_3[b'data'],dict_4[b'data'],dict_5[b'data']], axis=0)
```

```python
print(len(labels))
print(array.shape)
```

```python
array = pd.DataFrame(array)
mod_arr = array.assign(label = labels)
```

```python
print(mod_arr)
```

```python
train=mod_arr.sample(frac=0.8,random_state=200) #random state is a seed value
test=mod_arr.drop(train.index)

Actual_label_train=train['label']
Actual_label_test = test['label']

train = train.drop(['label'],axis=1)
test = test.drop(['label'],axis=1)
print(train)
print(test)
print(Actual_label_train)
print(Actual_label_test)
```

   # Train

```python
from sklearn.svm import SVC
clf = SVC(gamma='auto',kernel='linear')
clf.fit(train,Actual_label_train)
print(clf.predict(test))
predicted_label=clf.predict(test)
Accu_score = accuracy_score(Actual_label_test,predicted_label)
print("Accuracy score",Accu_score)
support_vec = clf.support_vectors_
print(support_vec)
```
