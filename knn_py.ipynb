```python
import csv
import math
from statistics import mode
import pandas as pd
import numpy
import sklearn
from scipy import stats as s
from sklearn.metrics import *

count=0
with open('train.csv') as csvfile:
    readCSV = csv.reader(csvfile)
    for r in readCSV:
        count=count+1

print(count)
tr=int(count*0.8)
va=int(count-tr)

print(tr)
print(va)

with open('train.csv') as csvfile:
    readCSV = csv.reader(csvfile)
    count=0
    trainingSet=[]
    validationSet=[]
    for r in readCSV:
        if count<tr:
            traini=[]
            for col in r:
                traini.append(int(col))
            trainingSet.append(traini)
            count=count+1;
        else:
            valid=[]
            for col in r:
                valid.append(int(col))
            validationSet.append(valid)
            count=count+1;

actual_label=[]
for trr in validationSet:
    actual_label.append(trr[0])
    
            
# for tr in trainingSet:
#     print(tr)
#     break
    
# for ts in validationSet:
#     print(ts)
#     break

    
def do_sort(label_dis):
    return(sorted(label_dis, key = lambda x: x[1]))  

euclid_dist=[]
s=[]
t=[]
cc=0
for i in range(0,va):
    s=[]
    s=validationSet[i]
    s1=s[1:]
    s1=numpy.array(s1)
    #print(s1)
    euclid_dis=[]
    for j in range(0,tr):
        t=[]
        t=trainingSet[j]
        label=t[0]
        t1=t[1:]
        t1=numpy.array(t1)
        dis=0
        label_dis=[]
#         for k in range(0,len(s1)):
#             dis=dis + ((s1[k]-t1[k])**2)
#         dis=math.sqrt(dis)
        dis = numpy.linalg.norm(s1 - t1)
        label_dis.append(label)
        label_dis.append(dis)
        euclid_dis.append(label_dis)
    cc=cc+1
    print(cc)
    euclid_dis_pair=do_sort(euclid_dis)
    euclid_dist.append(euclid_dis_pair)
    
        
c=0;
for dist in euclid_dist:
    print(dist)
    c=c+1
    print("printing size of dis")
    print(len(dist))
    print("printing..")
    break
        
        
manhattan_dist=[]
for i in range(0,va):
    s=[]
    s=validationSet[i]
    s1=s[1:]
    manhattan_dis=[]
    for j in range(0,tr):
        t=[]
        t=trainingSet[j]
        label=t[0]
        t1=t[1:]
        m_dis=0
        label_dis=[]
        for k in range(0,len(s1)):
            m_dis=m_dis+ abs(s1[k]-t1[k])
        label_dis.append(label)
        label_dis.append(m_dis)
        manhattan_dis.append(label_dis)
    manhattan_dis_pair = do_sort(manhattan_dis)
    manhattan_dist.append(manhattan_dis_pair)
    break
    
print("**************************MANHATTAN*****************************************")
for l in manhattan_dist:
    print(l)
    print("printing size of dis")
    print(len(l))
    print("printing")
    break


    

    
            
    


```
